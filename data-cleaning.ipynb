```python
import pandas as pd

raw_path = "raw/"
clean_path = "clean/"
```

```python
food_kg_df_raw = pd.read_csv('raw/nourish_public_FoodKG.csv').dropna(axis=1)
food_kg_df_raw.head().to_csv(clean_path + "nourish_public_FoodKG.csv", index=False)
food_kg_df_raw.columns.to_list()
```

```python
experimental_df_raw = pd.read_csv('raw/nourish_public_usda__branded_experimental.csv').dropna(axis=1)
experimental_df_raw.head().to_csv(clean_path + "nourish_public_usda__branded_experimental.csv", index=False)
experimental_df_raw.columns.to_list()
```

```python
product_df_raw = pd.read_csv('raw/nourish_public_usda__branded_food_product.csv').dropna(axis=1)
product_df_raw.head().to_csv(clean_path + "nourish_public_usda__branded_food_product.csv", index=False)
product_df_raw.columns.to_list()
```

```python
nutrients_df_raw = pd.read_csv('raw/nourish_public_usda_randed_food_nutrients.csv').dropna(axis=1)
nutrients_df_raw.head().to_csv(clean_path + "nourish_public_usda_randed_food_nutrients.csv", index=False)
nutrients_df_raw.columns.to_list()
```

```python
nutrients_master_df_raw = pd.read_csv('raw/nourish_public_usda_2022_nutrient_master.csv').dropna(axis=1, how='all')
nutrients_master_df_raw.head().to_csv(clean_path + "nourish_public_usda_2022_nutrient_master.csv", index=False)
nutrients_master_df_raw.columns.to_list()
```
