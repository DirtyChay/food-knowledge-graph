### combine_batches_with_single_run
This notebook combines the output of `read_batches` with the asynchronous LLM output generated by `process_products`.

```python
import os

from dotenv import load_dotenv

load_dotenv()  # Take environment variables from .env
import pandas as pd
```

```python
raw_data_path = "../../data/raw/usda_2022_food_branded_experimental_DESCRIPTION_ONLY.csv"

if os.path.exists(raw_data_path):
    raw_data = pd.read_csv(raw_data_path)
    print(len(raw_data))
    display(raw_data.head())
else:
    raise FileNotFoundError(f"File does not exist: {raw_data_path}")
```

```python
batch_path = "../batches/batch_results_all.csv"

if os.path.exists(batch_path):
    batch_results_df = pd.read_csv(batch_path)
    print(len(batch_results_df))
    display(batch_results_df.head())
else:
    raise FileNotFoundError(f"File does not exist: {batch_path}")
```

```python
batch_results_df = batch_results_df.merge(
    raw_data[['fdc_id', 'description']],
    on='fdc_id',
    how='left'
)
batch_results_df = batch_results_df.rename(columns={'mapped_ingredient': 'response'})
print(len(batch_results_df))
batch_results_df.head()
```

```python
single_run_df = pd.read_csv("../single_run/results_async.csv")
print(len(single_run_df))
single_run_df.head()
```

```python
combined_df = pd.concat([single_run_df, batch_results_df], ignore_index=True).drop_duplicates("fdc_id")
print(len(combined_df))
combined_df.head()
```

```python
print(len(combined_df) == len(raw_data))
```

```python
combined_df = combined_df.sort_values(by='fdc_id', ascending=True)
combined_df.to_csv("mapped_responses.csv", index=False)
```
