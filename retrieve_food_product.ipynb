```python
import os

import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine

load_dotenv()  # Take environment variables from .env
```

```python
# Credentials
host = os.getenv("POSTGRES_HOST")
port = os.getenv("POSTGRES_PORT")
database = os.getenv("POSTGRES_DATABASE")
user = os.getenv("POSTGRES_USER")
password = os.getenv("POSTGRES_PASSWORD")
# Create the connection engine
engine = create_engine(f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}")
# Test the connection
try:
    with engine.connect() as conn:
        print("✅ Connected successfully!")
except Exception as e:
    print("❌ Connection failed:", e)
```

```python
query = """
        SELECT fdc_id,
               ingredients
        FROM usda_2022_branded_food_product
        WHERE fdc_id IS NOT NULL
          AND ingredients IS NOT NULL LIMIT 500; \
        """
df = pd.read_sql(query, engine)
```

```python
df.to_csv("data/usda_2022_branded_food_product.csv", index=False)  # save entire database
```

```python
filtered_df = df[["fdc_id", "ingredients"]].sort_values(by=["id"])  # get only id and ingredients
filtered_df.head()
```

```python
filtered_df.to_csv("data/usda_2022_branded_food_product_ingredients.csv", index=False)  # save only id and ingredients
```
